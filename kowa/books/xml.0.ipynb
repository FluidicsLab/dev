```python
from lxml import etree

class KowaConfig:
    
    _items = {}

    def __str__(self):
        for key in self._items.keys():
            for mode in self._items[key].keys():
                print(f"{key}({mode}) {self._items[key][mode]}")
        return ""

    def update(self, key, mode, name, attr):        
        if key not in self._items.keys():
            self._items[key] = {}  
        if mode not in self._items[key].keys():
            self._items[key][mode] = []
        self._items[key][mode].append({ name: attr })

_config = KowaConfig()

kowa_filename = "../config/kowa.xml"

tree = etree.parse(kowa_filename)
if tree is not None:
    
    root = tree.getroot()

    for cat in root.xpath("./Group/Category[1]"):
        
        key = cat.get('Name')

        ifeat = cat.xpath(f"../Integer")
        ffeat = cat.xpath(f"../Float")
        efeat = cat.xpath(f"../Enumeration")
        sfeat = cat.xpath(f"../StringReg")

        if len(ifeat) > 0:
            for feat in ifeat:
                name = feat.get('Name')
                attr = {}
                _config.update(key, 'integer', name, attr)
        if len(ffeat) > 0:
            for feat in ffeat:
                name = feat.get('Name')
                attr = None
                _config.update(key, 'float', name, attr)
        if len(efeat) > 0:
            for feat in efeat:
                name = feat.get('Name')
                attr = { 'keys': feat.xpath('EnumEntry/@Name'), 'values': feat.xpath('EnumEntry/Value/text()') }
                _config.update(key, 'enumeration', name, attr)
        if len(sfeat) > 0:        
            for feat in sfeat:
                name = feat.get('Name')
                attr = None
                _config.update(key, 'string', name, attr)

print(_config)                
        
```
