```python

import pyexiv2
import os
import numpy as np

folder = r"C:\Develop\py\InspIOR-Vision-Pro-V0.9.2\projects\archive\250213_Team1"

extension = '.jpg'
prefix = '0_VLXT-650C_I_EF'

names = [name for name in os.listdir(folder) if name.endswith(extension) and name.startswith(prefix)]

indizes = np.array([int(name.replace(prefix, '').replace(extension,'')) for name in names]).argsort()
filenames = np.array([os.path.join(folder,name) for name in names])[indizes]

for filename in filenames[:3]:

    img = pyexiv2.Image(filename)
    shape = [img.get_pixel_height(), img.get_pixel_width()]
    xmp = img.read_xmp()
    img.close()

    print(f"{filename} {shape}")

    for key in xmp.keys():
        name = key.split(".")
        print(f"{key:>20} {name[-1]:>10}  {xmp[key]}")

```
