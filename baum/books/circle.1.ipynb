```python
import cv2
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy import ndimage

img = cv2.imread(r'C:\Data\0_VLXT-650C_I_EF1724912396928.jpg')
i8 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


#idx = np.argwhere(it == 200)

#p1 = idx.min(axis=0)[::-1]
#p2 = idx.max(axis=0)[::-1]

#_ = cv2.rectangle(it, p1, p2, (200,0,0), 2)

#print(p2 - p1)
```

```python
_, it = cv2.threshold(255-i8, thresh=120, maxval=200, type=cv2.THRESH_BINARY_INV)

sy = ndimage.gaussian_laplace(255-it, sigma=1)
shape = sy.shape

sy = sy.reshape(-1,2)
sy[np.argwhere(sy>0)] = 255
sy = sy.reshape(shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
for i,p in enumerate([sy, it]): 
    ax[i].imshow(p,cmap='grey')
for a in ax: 
    a.axis("off")
plt.show()
plt.close()
```
