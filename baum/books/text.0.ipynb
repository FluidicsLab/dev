```python
import os
import cv2
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

img = cv2.imread(r'..\media\testcard_color.png')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h,w,d = rgb.shape

fparam = {
    'font': cv2.FONT_HERSHEY_SIMPLEX,
    'scale': .8,
    'thickness': 2,
    'color': (201,0,0)
}

ftext = 'no image data'

fsize, _ = cv2.getTextSize(ftext, fparam['font'], fparam['scale'], fparam['thickness'])
fw, fh = fsize

x, y = int(w/2-fw/2), int(h/4-fh/2)

print(x,y)

cv2.putText(rgb, ftext, (x,y), fparam['font'], fparam['scale'], fparam['color'], fparam['thickness'])

imgz = [rgb]
N = len(imgz) 

aspect = h/w


fig, ax = plt.subplots(1, N, figsize=( 10, aspect * 10 * N))

if not isinstance(ax, np.ndarray): 
    ax = [ax]

for i,p in enumerate(imgz): 
    ax[i].imshow(p,cmap='grey')
for a in ax: 
    a.axis("off")
plt.show()
plt.close()
```
