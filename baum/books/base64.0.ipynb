```python
import os
import base64
import cv2
import urllib.parse

def svg2b64(folder, filenames):
    for filename in filenames:
        filename = os.path.join(folder, filename)
        with open(filename, "rb") as f:
            b64 = base64.b64encode(f.read()).decode()
        with open(f"{filename}.b64", "w") as f:
            f.write(b64)

def img2b64(folder, filenames):
    for filename in filenames:
        filename = os.path.join(folder, filename)
        img = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
        b64 = base64.b64encode(cv2.imencode('.png', img)[1]).decode()
        with open(f"{filename}.b64", "w") as f:
            f.write(b64)

def svg2url(folder, filenames):
    for filename in filenames:
        filename = os.path.join(folder, filename)
        with open(filename, "r") as f:
            url = f.read()
        url = url.replace('\n', ' ')
        url = url.replace('<', '%3C')
        url = url.replace('>', '%3E')
        url = url.replace('{', '%7B')
        url = url.replace('}', '%7D')
        url = url.replace(';', '%3B')
        url = url.replace('#', '%23')        
        url = f"url(\"data:image/svg+xml,{url}\");"
        with open(f"{filename}.url", "w") as f:
            f.write(url)
```
