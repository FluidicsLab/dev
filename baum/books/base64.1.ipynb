```python

import base64
import time

from PIL import Image
from io import BytesIO

filename = r"C:\Data\base64\0_VLXT-650C_I_EF1755203914641.jpg"

import cv2
from turbojpeg import TurboJPEG, TJFLAG_FASTUPSAMPLE, TJFLAG_FASTDCT
tj = TurboJPEG(r"C:\Develop\py\FluidicsLab\Drafts\baum\lib\libjpeg-turbo\bin\turbojpeg.dll")

image = cv2.imread(filename,cv2.IMREAD_UNCHANGED)

thumb = image

#thumb = cv2.resize(image,(0,0), fx=0.4, fy=0.4)
thumb = cv2.cvtColor(thumb, cv2.COLOR_BGR2RGB)
```

```python
t = time.time()

_, buffer = cv2.imencode('.jpg', thumb, [int(cv2.IMWRITE_JPEG_QUALITY), 87])

#_, buffer = cv2.imencode('.png', thumb, [int(cv2.IMWRITE_PNG_COMPRESSION), 3])
#_, buffer = cv2.imencode(".webp", thumb, [int(cv2.IMWRITE_WEBP_QUALITY), 80])

print(len(buffer), time.time()-t)
```

```python
t = time.time()

buffer = tj.encode(thumb)

print(len(buffer), time.time()-t)
```

```python
t = time.time()

enc = base64.b64encode(buffer)
b64 = enc.decode()

print(len(b64), time.time()-t)
```

```python

im = Image.open(BytesIO(buffer))
im.save(f"{filename}.jpg", 'JPEG')
```
