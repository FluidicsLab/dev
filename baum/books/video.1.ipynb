```python
import numpy as np
import cv2
import os

folder = os.path.join("C:\\Data\\240915")

extension = '.jpg'
prefix = '0_VLXT-650C_I_EF'#'1_VLXT-81C_I'

names = [name for name in os.listdir(folder) if name.endswith(extension) and name.startswith(prefix)]
indizes = np.array([int(name.replace(prefix, '').replace(extension,'')) for name in names]).argsort()
filenames = np.array([os.path.join(folder,name) for name in names])[indizes]

image = cv2.imread(filenames[0], cv2.IMREAD_UNCHANGED)
size = (image.shape[:2])[::-1]

scale = 0.5
size = (int(size[0]*scale), int(size[1]*scale))

size
```

```python
filename = f"{folder}\{prefix}.avi"

codec = "DIVX"
fps = 2

fourcc = cv2.VideoWriter_fourcc(*(codec))

video:cv2.VideoWriter = cv2.VideoWriter()
video.open(filename, fourcc, fps, size, True)

for filename in filenames:
    frame = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
    frame = cv2.resize(frame, size)
    video.write(frame)

video.release()
```
