```python
from multiprocessing import shared_memory

import numpy as np

name = "EcatSeverityController"
len_ = 16
size = len_ * 8

ctrl = None

rc = 0
try:
    ctrl = shared_memory.SharedMemory(name=name, create=True, size=size)
    rc = 1
except FileExistsError:
    ctrl = shared_memory.SharedMemory(name=name, create=False, size=size)
    rc = 2
except Exception:
    ctrl = None
    rc = 3
    
print(rc)

severity = np.ndarray((len_,), dtype=np.int64, buffer=ctrl.buf)

current = severity.copy()

for i in range(len(current)):
    print(f"{i:02d}: {bin(severity[i]).replace('0b','').zfill(8)} <-- {bin(current[i]).replace('0b','').zfill(8)}")
    severity[i] = 0

severity[0] = 1

#ctrl.close()
ctrl.unlink() 
```
