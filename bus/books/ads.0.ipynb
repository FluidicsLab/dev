```python
import pyads
import time

def set_encoder_to_zero(plc):
    
    # enable drive
    plc.write_by_name('MAIN.bEnableDrive', True, pyads.PLCTYPE_BOOL)
    time.sleep(0.1)
    
    # enable regulator
    plc.write_by_name('MAIN.bRegulatorOn', True, pyads.PLCTYPE_BOOL)
    time.sleep(0.1)
    
    # current position
    current_pos = plc.read_by_name('MAIN.fActPosition', pyads.PLCTYPE_LREAL)
    print(f"current position {current_pos}")
    
    # set position
    plc.write_by_name('MAIN.nSetPosition', 0, pyads.PLCTYPE_DINT)
    
    plc.write_by_name('MAIN.bSetPosCmd', True, pyads.PLCTYPE_BOOL)
    time.sleep(0.05)
    plc.write_by_name('MAIN.bSetPosCmd', False, pyads.PLCTYPE_BOOL)
    
    time.sleep(0.2)
    new_pos = plc.read_by_name('MAIN.fActPosition', pyads.PLCTYPE_LREAL)
    print(f"new position {new_pos}")

plc = None
try:

    plc = pyads.Connection(
        ams_net_id='169.254.180.216.5.1',
        ams_net_port=851
        )  # AMS-NetId, Port
    plc.open()

    print(plc.read_state())
    print(plc.get_all_symbols())

    # set_encoder_to_zero(plc)

except Exception as e:
    print(f"ERROR: {e}")

finally:
    if plc is not None:
        plc.close()
```
