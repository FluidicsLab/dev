```python
import ctypes

class PositionEx:

    UINT32_MAX = 4_294_967_295
    UINT32_BIT = 32

    @staticmethod
    def split(value, bits):            
        value = bin(value)[2:].zfill(32)
        return [int(value[:bits].zfill(32),2), int(value[bits:].zfill(32),2)]
    
    @staticmethod
    def merge(value, bits):
        return PositionEx.value(int("".join([bin(value[0])[2:].zfill(bits), bin(value[1])[2:].zfill(PositionEx.UINT32_BIT-bits)]), 2))

    @staticmethod
    def value(value):
        return PositionEx.UINT32_MAX + value if value < 0 else value

i = 3793892265
bits = 15

m = PositionEx.merge([28945, 0],bits)

print(m)
```
