```python

class AM8111ProfilePosition:

    UINT32_MAX = 2**32 - 1
    UINT64_MAX = 2**64 - 1

    @staticmethod
    def split(value, bits, range=32):            
        value = bin(value)[2:].zfill(range)
        return [
            int(value[:bits].zfill(range),2), 
            int(value[bits:].zfill(range),2)
            ]
    
    @staticmethod
    def merge(value, bits, range=32):
        return AM8111ProfilePosition.value(
            int("".join([
            bin(value[0])[2:].zfill(bits), 
            bin(value[1])[2:].zfill(range-bits)
            ]), 2), range)
    
    @staticmethod
    def value(value, range=32):
        return value
        if range == 32:
            return AM8111ProfilePosition.UINT32_MAX + value if value < 0 else value
        else:
            return AM8111ProfilePosition.UINT64_MAX + value if value < 0 else value
        
    @staticmethod
    def compare(raw, value, tol=[None, None], bits=32, range=32):
        raw = AM8111ProfilePosition.split(raw, bits)
        win = [abs(raw[0]-value[0]) <= tol[0] if tol[0] is not None else True, abs(raw[1]-value[1]) <= tol[1] if tol[1] is not None else True]
        if raw[0] < value[0]:
            return -1, win
        elif raw[0] > value[0]:
            return +1, win
        else:
            if raw[1] < value[1]:
                return -1, win
            elif raw[1] > value[1]:
                return +1, win
        return 0, win
    
    @staticmethod
    def sub(a, b, bits):
        return AM8111ProfilePosition.merge(a, bits) - AM8111ProfilePosition.merge(b, bits)

    @staticmethod
    def add(a, b, bits):
        return AM8111ProfilePosition.merge(a, bits) + AM8111ProfilePosition.merge(b, bits)

import ctypes

imx = 2**32//2-1

bits = 16

off = AM8111ProfilePosition.merge([57_548,0], bits)

pos = AM8111ProfilePosition.merge([18000,0], bits)
dst = AM8111ProfilePosition.merge([1619,0], bits)

mx = AM8111ProfilePosition.merge([19_935,0], bits)

print(f"{pos:15d}")
print(f"{dst:15d}")

print(f"{pos-dst:15d}")

print(f"{mx:15d}")

print(f"{off:15d}")

AM8111ProfilePosition.split(1306460160, bits)
```

```python
from functools import reduce

def find(data,path):
    try:
        return reduce(lambda acc,i: acc[i], path.split('.'), data)
    except:
        return None
    
find({ 'value': { 'value': 1 } }, 'value.value')
```
