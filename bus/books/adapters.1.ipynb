```python
import ifaddr, getmac

import subprocess

def get_mac_address(ip_address):
    arp_command = ['arp', '-a', ip_address]
    output = subprocess.check_output(arp_command).decode()
    mac_address = output.split()[3]
    return mac_address

def get_ip_address(ips):
    for ip in ips:
        if ip.network_prefix in [16,24]:
            return ip
    return None

adapters = ifaddr.get_adapters(include_unconfigured=False)

for adapter in adapters:
    
    print(f"{adapter.nice_name}")
    
    ip = get_ip_address(adapter.ips)
    print(ip)

    #mac = get_mac_address(ip.ip)
    #print(mac)    
    

```
