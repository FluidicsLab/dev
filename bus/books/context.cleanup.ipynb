```python

import os
import json

root = r'C:\\Users\\Aramco\\.node-red\\t\\context\\'
#root = r'C:\\Users\\Aramco\\.node-red\\d4\\context\\'

remove = True

def loop_(folder, rc):
    for root, dirs, files in os.walk(folder):
        for filename in files:
            rc.append(os.path.join(root, filename))
        for dirname in dirs:
            loop_(os.path.join(root, dirname), rc)
    
filenames = []
loop_(root, filenames)

def is_json(data):
    try:
        json.loads(data)
    except:
        return False
    return True

filenames = list(set(filenames))

for filename in filenames:
    data = None
    valid = False
    with open(filename, "r") as f:
        valid = is_json(f.read())
    if not valid:
        print(f"invalid context file {filename} - remove")
        if remove:
            os.remove(filename)

print('done')
    
```
