```python
import psutil

print(f"version {psutil.__version__}")

name = "node.exe"

pids = [p.info['pid'] for p in psutil.process_iter(['pid', 'name']) if p.name() == name]

for pid in pids:

    p = psutil.Process(pid=pid)
    with p.oneshot():
        print(f"{p.name()} {p.pid} {p.nice()} {p.ionice()} {p.cpu_times()}")

psutil.IOPRIO_HIGH, psutil.IOPRIO_NORMAL, psutil.IOPRIO_LOW, psutil.IOPRIO_VERYLOW
```
