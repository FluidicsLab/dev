```python

def crc_(data: list[int]):
    data = bytearray(data)
    crc = 0xFFFF
    for b in data:
        crc ^= b
        for _ in range(0,8):
            bcarry = crc & 0x0001
            crc >>= 1
            if bcarry: 
                crc ^= 0xa001
    rc = hex(crc).replace('0x','').zfill(4)
    crc = [rc[2:4],rc[0:2]]  
    return [int(crc[0],16),int(crc[1],16)]    

def value_(s):
    rc = s.split(" ")
    rc = int(rc[3],16)<<8|int(rc[4],16)    
    return rc
```
